**useful for making plots**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100) #evenly spaced values from 0 to 10 100 values
y = np.sin(x)
z = np.cos(x)
print(x)
```

**plotting the data**

```python
#sine graph
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding labels and titles :
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine value')
plt.title('Sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
#Dotted lines and other patterns
plt.plot(x,y,'r.') #can use r+ for +, r* for * and so on | r is for red, g for
#green
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r.')
plt.show()
```

**Bar plot**




```python
fig = plt.figure() #CREATES empty plot
ax = fig.add_axes([0,0,1,1]) #adding axes to the empty figure
lang = ['eng','hin','ben','spa','fre']
people = [100,90,50,20,30]
ax.bar(lang,people)
plt.xlabel('languages')
plt.ylabel('no.of people')
plt.plot()
plt.show()
```

**pie chart** 

```python
fig = plt.figure() #CREATES empty plot
ax = fig.add_axes([0,0,1,1]) #adding axes to the empty figure
lang = ['eng','hin','ben','spa','fre']
people = [100,90,50,20,30]
ax.pie(people,labels = lang, autopct='%1.1f%%') #autopct for decimal vals
plt.show()
```

**Scatter Plot**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z, color = 'r')
plt.show()
```

**3D scatter plot**

```python
fig2 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,cmap='Blues')
plt.show()
```
