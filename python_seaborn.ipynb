**It i s a Data visualization library**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#seaborn has some built in datasets : eg - total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme()
```

```python
#visualize the ds
sns.relplot(data=tips, x='total_bill', y='tip', col='time',hue='smoker',style='smoker', size='size')
#differentiates stuff by its own. we dont need to mention much
```

```python
#load the IRIS dataset 
iris = sns.load_dataset('iris') #iris flower
iris.head()
iris.tail()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris) 
#plot differentiated based on species (in hue)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue ='species', data=iris)
```

Housing dataset

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
#now storing in pandas df
house = pd.DataFrame(housing.data,columns=housing.feature_names) #convert
#numpy to pandas df
house['PRICE'] = housing.target
house.head()
#predict price of a house
```

Distribution plot 

```python
sns.displot(house['PRICE']) #x axis-> 1000 dollars ,, count - no of houses
#distplot gives the range in which the data is distributed
```

**Correlation**

```python
correlation = house.corr()
```

```python
#HEAT MAP FOR CORRELATION
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True,fmt='.1f', annot=True, annot_kws={'size':8},
            cmap='Reds')
```
